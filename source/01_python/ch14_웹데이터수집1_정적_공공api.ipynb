````python
# div.head_info > span.value(find함수를 쓰면 복잡해짐)

prices = []
headinfos = soup.find_all('div', class_='head_info')
for headinfo in headinfos:
    price = headinfo.find('span', class_='value')
    # print(price.text.replace(',',''))
    prices.append(float(''.join(price.text.split(','))))
print(prices)
````

````python
# div.head_info > span.value(select함수)

prices = soup.select('div.head_info > span.value')
print([float(p.text.replace(',','')) for p in prices])
````

````python
titles = soup.select('h3.h_lst > span.blind')
[t.text for t in titles]
````

````python
# div.head_info > span.blind : 이건 페이지 내 '하락'을 가져오게 됨
# div.head_info > span.txt_krw > span.blind : 이건 달러를 가져오지 않음
# div.head_info > span > span.blind 그래서 이런식으로 가져와야함

units = soup.select('div.head_info > span > span.blind')
units = [unit.text for unit in units]
units.insert(7,'') # 7번째에 '' 추가. 나머지는 다 12개씩 있는데 이것만 중국쪽이 비어서
units
````

````python
statuses = soup.select('div.head_info > span.blind')
[status.text for status in statuses]

# for idx in range(len(statuses)):
#     print(statuses[idx].text, end='\t')
````

````python
# 있어보이는 for문

for title, price, unit, status in zip(titles, prices, units, statuses):
    print(f'{title.text} : {price.text}{unit} - {status.text}')
````

````python
# 1. 다음 뉴스 검색(원하는 키워드, 원하는 페이지를 가져오기)

import requests
from bs4 import BeautifulSoup
import pandas as pd

def collect_daum_news_list(keyword, page):
    'keyword로 page에 다음 뉴스 검색한 결과를 출력 → list를 return'
    # url = f'https://search.daum.net/search?w=news&nil_search=btn&DA=PGD&enc=utf8&cluster=y&cluster_page=1&q={keyword}&p={page}'
    # print(url) # 출력이 제대로 되는지 확인
    # response = requests.get(url) url이 너무 길어서 수정할거임
    url = f'https://search.daum.net/search?w=news&nil_search=btn&DA=PGD&enc=utf8'
    params = {'q':keyword, 'p':page} # 이렇게 따로 보낼수도 있음
    response = requests.get(url, params=params)
    soup = BeautifulSoup(response.text, 'html.parser')
    
    items_find_list = [] # 검색할 결과를 담는 리스트
    item_titles = soup.select('div.item-title > strong.tit-g > a')
    for idx, item in enumerate(item_titles):
        items_find_list.append([(page-1)*10+idx, item.text, item.attrs['href']])
    return items_find_list
````

````python
collect_daum_news_list('AI',1)
````

````python
import time

keywords = ['청바지', '동대문']
result0 = [] # 청바지 1~3페이지까지 검색한 결과를 담을 list
result1 = [] # 동대문 1~3페이지까지 검색한 결과를 담을 list
pages = 3

for i, keyword in enumerate(keywords):
    print(f'==={i+1}번째 검색어 {keyword} 검색 결과 수집 중입니다===')
    for page in range(1, pages+1):
        if i==0:
            result0.extend(collect_daum_news_list(keyword, page)) # append를 쓰면 3차원이 되어버림. 그럼 데이터프레임으로 못한다
        elif i==1:
            result1.extend(collect_daum_news_list(keyword, page))
        time.sleep(0.5) # 중간중간 재우는것
````

````python
result0_df = pd.DataFrame(result0, columns=['no','title','link'])
result1_df = pd.DataFrame(result1, columns=['no','title','link'])
result0_df.head()
````

````python
import requests
import pandas as pd
import json
from dotenv import load_dotenv
import os
load_dotenv()

client_id = os.getenv('CLIENT_ID')
client_secret = os.getenv('CLIENT_SECRET')
headers = {'X-Naver-Client-Id': client_id, 'X-Naver-Client-Secret' : client_secret}
query = '청바지'
url = 'https://openapi.naver.com/v1/search/image'
params = {'query':query, 'display':100}
response = requests.get(url, params=params, headers=headers)

items = response.json()['items']
items_list = []
for item in items:
    title = item.get('title')
    link = item.get('link')
    sizeheight = item.get('sizeheight')
    sizewidth = item.get('sizewidth')
    items_list.append([title, link, sizeheight, sizewidth])
    
df = pd.DataFrame(items_list, columns=['title', 'link', 'sizeheight', 'sizewidth'])
df.head()
````

````python
df.to_csv('data/ch14_청바지.csv', index=False)
````

````python
def get_image_list(query):
    'query로 검색한 이미지정보(정보,링크,썸네일,크기) 100건의 데이터 프레임을 return(방법1)'
    import requests
    import pandas as pd
    import json
    from dotenv import load_dotenv
    import os
    load_dotenv()

    client_id = os.getenv('CLIENT_ID')
    client_secret = os.getenv('CLIENT_SECRET')
    headers = {'X-Naver-Client-Id': client_id, 'X-Naver-Client-Secret' : client_secret}
    query = '청바지'
    url = 'https://openapi.naver.com/v1/search/image'
    params = {'query':query, 'display':100}
    response = requests.get(url, params=params, headers=headers)

    items = response.json()['items']
    
    items_list = []
    for item in items:
        link = item.get('link')
        thumbnail = item.get('thumbnail')
        items_list.append({
            '제목':item.get('title'),
            '링크':link,
            '썸네일': thumbnail,
            'sizeheight': int(item.get('sizeheight')),
            'sizewidth':  int(item.get('sizewidth')),
        })
        
    return pd.DataFrame(items_list)
df = get_image_list("청바지")
df.to_csv('data/ch14_청바지.csv')
df.sample()
````

````python
def save_image(attr, idx, link, query):
    'link의 이미지를 image/attr_idx_query.확장자로 local에 저장'
    file_extension = link.split('.')[-1] # 확장자
    quote_index = file_extension.find('?') # 확장자 뒤에 ?가 있는 위치를 반환
    if quote_index != -1:
        file_extension = file_extension[:quote_index]
    img = requests.get(link).content # 바이너리
    with open(f'image/{attr}_{idx+1:03}_{query}.{file_extension}', 'wb') as f:
        f.write(img)
    
    
save_image('메인', 0, df.loc[0, '링크'], '청바지')
````

````python
# 위의 함수를 적용시켜서 저장

def get_image_list_save_image(query):
    'query로 검색한 이미지정보(정보,링크,썸네일,크기) 100건의 데이터 프레임을 return(방법1)'
    import requests
    import json
    import pandas as pd
    from dotenv import load_dotenv
    import os
    load_dotenv()

    client_id = os.getenv('CLIENT_ID')
    client_secret = os.getenv('CLIENT_SECRET')
    headers = {'X-Naver-Client-Id': client_id, 'X-Naver-Client-Secret' : client_secret}
    query = '청바지'
    url = 'https://openapi.naver.com/v1/search/image'
    params = {'query':query, 'display':100}
    response = requests.get(url, params=params, headers=headers)
    items = response.json()['items']
    
    items_list = [] # 정보가 담길 리스트
    for idx, item in enumerate(items):
        link = item.get('link')
        thumbnail = item.get('thumbnail')
        items_list.append({
            '제목':item.get('title'),
            '링크':link,
            '썸네일': thumbnail,
            'sizeheight': int(item.get('sizeheight')),
            'sizewidth':  int(item.get('sizewidth')),
        })
        
        #이미지 파일 저장
        save_image('메인',idx,link,query)
        save_image('썸네일', idx, thumbnail, query)
        if (idx%20 == 0) & (idx!=0):
            print(f'= = = {idx}% 진행중 = = =')
    print('= = = 완료 = = =')
    return pd.DataFrame(items_list)
df = get_image_list_save_image("청바지")
df.to_csv('data/ch14_청바지.csv')
df.sample()
````

## 2.3 XML 웹데이터 수집

### 1) 전국 날씨 RSS를 BeautifulSoup을 이용한 크롤링

- 구글에 '기상청 RSS' 검색한 첫번째 사이트 이동

````python
import requests
from urllib.request import urlopen
from bs4 import BeautifulSoup
import numpy as np
import pandas as pd

items_list = []
url = 'http://www.kma.go.kr/repositary/xml/fct/mon/img/fct_mon1rss_108_20251106.xml'

# 방법 1
target = requests.get(url)
soup = BeautifulSoup(target.content, 'xml') # 다른 xml parser방식은 pip install lxml해야함.

# 방법 2
# target = urlopen(url)
# soup = BeautifulSoup(target, 'xml')

local_tas = soup.select('local_ta')
for local_ta in local_tas:
    local_name = local_ta.select_one('local_ta_name').text
    week1_nomalYear = local_ta.select_one('week1_local_ta_normalYear').text
    week1_similarRange = local_ta.select_one('week1_local_ta_similarRange').text
    week1_minVal = local_ta.select_one('week1_local_ta_minVal').text
    week1_similarVal = local_ta.select_one('week1_local_ta_similarVal').text
    week1_maxVal = local_ta.select_one('week1_local_ta_maxVal').text
    
    items_list.append({'지역':local_name,
                      '1주 평년기온':week1_nomalYear,
                      '1주 기온범위':week1_similarRange,
                      '1주 낮을확률':week1_minVal,
                      '1주 비슷할확률':week1_similarVal,
                      '1주 높을확률':week1_maxVal})

    df = pd.DataFrame(items_list)
df['1주 평년기온'] = df['1주 평년기온'].astype(np.float64)
df['1주 낮을확률'] = df['1주 낮을확률'].astype(np.int16)
df['1주 비슷할확률'] = df['1주 비슷할확률'].astype(np.int16)
df['1주 높을확률'] = df['1주 높을확률'].astype(np.int16)

df.head()
````

````python
# 평소하던 방법 1 - 버스ID 가져오기

import requests
from bs4 import BeautifulSoup
# urlretrieve(url, 저장경로) : url의 파일을 저장경로에 저장(한글 들어가면 안됨)
from urllib.request import urlretrieve 
from urllib.parse import quote

# busNum = quote('마포06')
busNum = '162'
key = os.getenv('KEY')
url1 = f'http://ws.bus.go.kr/api/rest/busRouteInfo/getBusRouteList?ServiceKey={key}&strSrch={busNum}'
response = requests.get(url1)
soup = BeautifulSoup(response.content, 'xml')

for item in soup.select('itemList'):
    busRouteNm = item.select_one('busRouteNm').text
    if busRouteNm == busNum:
        busRouteId = item.select_one('busRouteId').text
        break;

print('busRouteId =', busRouteId)
````

````python
# STEP 2 : 버스ID로 버스정류장목록 받아오기(정류장명, 정류장ID, 경도, 위도)
# 서울특별시_노선정보조회 서비스 - 4번기능 15page(getStaionsByRouteList) 이용
````

````python
import pandas as pd

url2 = f'http://ws.bus.go.kr/api/rest/busRouteInfo/getStaionByRoute?ServiceKey={key}&busRouteId={busRouteId}'
response = requests.get(url2)
soup = BeautifulSoup(response.content, 'xml')
itemLists = soup.select('itemList')
print(f'{busNum}번 정류장 갯수 : {len(itemLists)}')

bus_station = []
for itemList in itemLists:
    stationNm = itemList.find('stationNm').text # 정류소 이름
    station   = itemList.find('station').text   # 정류소ID
    gpsX      = itemList.find('gpsX').text      # 경도
    gpsY      = itemList.find('gpsY').text      # 위도    
    bus_station.append([stationNm, station, gpsX, gpsY])
    
df_station = pd.DataFrame(bus_station, columns=['정류소명','id','경도','위도'])
df_station
````

````python
# STEP 3 : 버스ID로 실시간위치정보를 받아오기(차량번호, 혼잡도, 최종정류장ID, 다음정류장ID, 도착소요시간)
# 서울특별시_버스위치정보조회 서비스 - 2번기능 8page (getBusPosByRtidList)이용
````

````python
url3 = f'http://ws.bus.go.kr/api/rest/buspos/getBusPosByRtid?serviceKey={key}&busRouteId={busRouteId}'
response = requests.get(url3)
soup = BeautifulSoup(response.content, 'xml')
itemLists = soup.select('itemList')
print(f'운행중인 버스는 {len(itemLists)}대 입니다')

bus_position = []
for itemList in itemLists:
    plainNo = itemList.select_one('plainNo').text # 차량번호
    #혼잡도(0 : 없음, 3 : 여유, 4 : 보통, 5 : 혼잡, 6 : 매우혼잡)
    congetion = itemList.select_one('congetion').text # 혼잡도
    congetion = '없음' if congetion=='0' \
           else '여유' if congetion=='3' \
           else '보통' if congetion=='4' \
           else '혼잡' if congetion=='5' \
           else '매우혼잡'
    bus_position.append([plainNo, congetion])
    
df_position = pd.DataFrame(bus_position, columns=['차량번호','혼잡도'])
df_position
````
