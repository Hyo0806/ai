````python
# 방법1 : url에 한글 포함되어도 상관없음

import requests
from bs4 import BeautifulSoup
url = 'https://finance.naver.com/marketindex/'
response = requests.get(url) 

# response는 편지지가 든 편지봉투. response.text(content)는 편지지
#response.status_code # 200이 뜨니까 url 제대로 가져온것
#response.text / response.content.decode('cp949')

soup = BeautifulSoup(response.text, 'html.parser')
````

````python
# 방법2 : url에 한글이 포함될시 추가적인 방법 필요

from urllib.request import urlopen
url = 'https://finance.naver.com/marketindex/'
response = urlopen(url)
response.status
# response.read().decode('cp949')

soup = BeautifulSoup(response, 'html.parser')
````

````python
# div.head_info > span.value(find함수를 쓰면 복잡해짐)

prices = []
headinfos = soup.find_all('div', class_='head_info')
for headinfo in headinfos:
    price = headinfo.find('span', class_='value')
    # print(price.text.replace(',',''))
    prices.append(float(''.join(price.text.split(','))))
print(prices)
````

````python
# div.head_info > span.value(select함수)

prices = soup.select('div.head_info > span.value')
print([float(p.text.replace(',','')) for p in prices])
````

````python
titles = soup.select('h3.h_lst > span.blind')
[t.text for t in titles]
````

````python
# div.head_info > span.blind : 이건 페이지 내 '하락'을 가져오게 됨
# div.head_info > span.txt_krw > span.blind : 이건 달러를 가져오지 않음
# div.head_info > span > span.blind 그래서 이런식으로 가져와야함

units = soup.select('div.head_info > span > span.blind')
units = [unit.text for unit in units]
units.insert(7,'') # 7번째에 '' 추가. 나머지는 다 12개씩 있는데 이것만 중국쪽이 비어서
units
````

````python
statuses = soup.select('div.head_info > span.blind')
[status.text for status in statuses]

# for idx in range(len(statuses)):
#     print(statuses[idx].text, end='\t')
````

````python
len(titles), len(prices), len(units), len(statuses)
````

````python
# 없어보이는 for문

for idx in range(len(titles)):
    print(f'{titles[idx].text} : {prices[idx].text}{units[idx]} - {statuses[idx].text}')
````

````python
# 있어보이는 for문

for title, price, unit, status in zip(titles, prices, units, statuses):
    print(f'{title.text} : {price.text}{unit} - {status.text}')
````

````python
# 1. 다음 뉴스 검색(원하는 키워드, 원하는 페이지를 가져오기)

import requests
from bs4 import BeautifulSoup
import pandas as pd

def collect_daum_news_list(keyword, page):
    'keyword로 page에 다음 뉴스 검색한 결과를 출력 → list를 return'
    # url = f'https://search.daum.net/search?w=news&nil_search=btn&DA=PGD&enc=utf8&cluster=y&cluster_page=1&q={keyword}&p={page}'
    # print(url) # 출력이 제대로 되는지 확인
    # response = requests.get(url) url이 너무 길어서 수정할거임
    url = f'https://search.daum.net/search?w=news&nil_search=btn&DA=PGD&enc=utf8'
    params = {'q':keyword, 'p':page} # 이렇게 따로 보낼수도 있음
    response = requests.get(url, params=params)
    soup = BeautifulSoup(response.text, 'html.parser')
    
    items_find_list = [] # 검색할 결과를 담는 리스트
    item_titles = soup.select('div.item-title > strong.tit-g > a')
    for idx, item in enumerate(item_titles):
        items_find_list.append([(page-1)*10+idx, item.text, item.attrs['href']])
    return items_find_list
````

````python
import time

keywords = ['청바지', '동대문']
result0 = [] # 청바지 1~3페이지까지 검색한 결과를 담을 list
result1 = [] # 동대문 1~3페이지까지 검색한 결과를 담을 list
pages = 3

for i, keyword in enumerate(keywords):
    print(f'==={i+1}번째 검색어 {keyword} 검색 결과 수집 중입니다===')
    for page in range(1, pages+1):
        if i==0:
            result0.extend(collect_daum_news_list(keyword, page)) # append를 쓰면 3차원이 되어버림. 그럼 데이터프레임으로 못한다
        elif i==1:
            result1.extend(collect_daum_news_list(keyword, page))
        time.sleep(0.5) # 중간중간 재우는것
````

````python
result0_df = pd.DataFrame(result0, columns=['no','title','link'])
result1_df = pd.DataFrame(result1, columns=['no','title','link'])
result0_df.head()
````

````python
result0_df.to_csv(f'data/ch14_{keywords[0]}.csv', index=False)
result1_df.to_csv(f'data/ch14_{keywords[1]}.csv', index=False)
````

````python
# 방법 2 : 한글 인코딩 처리, user-agent추가

from urllib.request import urlopen, Request
from urllib.parse import quote # 한글을 encoding 처리하기 위함

# 한글 인코딩 처리
word = quote('비트코인')
url = 'https://search.daum.net/search?w=news&nil_search=btn&DA=PGD&enc=utf8&q=' + word
print(url)

# user-agent추가
headers = {'User-Agent':'Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/142.0.0.0 Safari/537.36'}
request = Request(url, headers=headers)
response = urlopen(request)
soup = BeautifulSoup(response, 'html.parser')
````
