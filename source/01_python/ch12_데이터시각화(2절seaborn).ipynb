## 2.3 seaborn 함수로 그래프 그리기

### 1) 관계형 그래프

- scatterplot, lineplot, relplot(서브플롯제공) : x가 수치형 데이터

```python
# 종별 다른색상과 다른 스타일로 scatterplot

sns.scatterplot(data=iris, x='petal_length', y='petal_width', hue='species', style='species')
plt.show()
```

```python
# 종별 다른색상과 다른 스타일로 lineplot

sns.lineplot(data=iris, x='petal_length', y='petal_width', hue='species', style='species')
plt.show()
```

```python
# 종별 lineplot을 서브플롯으로 

sns.relplot(data=iris, x='petal_length', y='petal_width', hue='species', 
            kind='line', # 서브플롯으로 그려질 그래프 종류(기본값 : scatter)
           col='species', # col단위로 서브플롯이 표현
           col_wrap=3, # 한줄에 표현될 서브플롯 수 (안쓰면 한줄에 다나옴)
           # facet_kws={'sharey':False} # 축공유
           )
plt.show()
```

```python
# 종별 scatterplot을 서브플롯으로 

g = sns.relplot(data=iris, x='petal_length', y='petal_width', hue='species', 
            kind='scatter', # 서브플롯으로 그려질 그래프 종류(기본값 : scatter)
            col='species', # col단위로 서브플롯이 표현
            col_wrap=2, # 한줄에 표현될 서브플롯 수 (안쓰면 한줄에 다나옴)
            # facet_kws={'sharey':False} # 축공유
           )

# 크기조정
g.fig.set_size_inches(10,3)
plt.show()
```

```python
# ax[0]에는 scatter, ax[1]에는 line

fig, ax = plt.subplots(ncols=2)
sns.scatterplot(data=iris, x='petal_length', y='petal_width', hue='species', ax=ax[0])
sns.lineplot(data=iris, x='petal_length', y='petal_width', hue='species', ax=ax[1])

plt.show()
```

### 2) 범주형 그래프

- 점도표 : stripplot(무작위로 흩뿌림), swarmplot(겹치지 않음/대신 느림)
- 막대그래프 : barplot, pointplot(막대대신 marker가 있는 선)
- countplot(bar는 y의 평균값 높이, count는 x의 빈도수 높이)
- boxplot, violinplot, boxenplot → catplot(서브플롯)

```python
# x축이 범주형일때의 산점도

sns.stripplot(data=iris, x='species', y='petal_width', hue='species') # palette='twilight_shifted'
plt.show()
```

```python
# x축이 범주형일때의 산점도(중첩되지 않게)

sns.swarmplot(data=iris, x='species', y='petal_width', hue='species')
plt.show()
```

```python
# 내림차순으로 정렬된 iris 범주형 산점도

sns.swarmplot(data=iris.sort_values('petal_width', ascending=False), x='species', y='petal_width', hue='species')
plt.show()
```

```python
# x축의 순서를 임의로 정할 수 있음

sns.swarmplot(data=iris, x='species', y='petal_width', hue='species',
             order=['versicolor','setosa','virginica'])
plt.show()
```

```python
# barplot : 각 범주별 평균값과 그에 대한 신뢰구간(신뢰구간제외 : errorbar=None)

sns.barplot(data=iris, x='species', y='sepal_width',
           errorbar=None)
sns.lineplot(data=iris, x='species', y='sepal_width', color='k')
plt.show()
```

```python
# pointplot : 각 범주별 평균값과 그에 대한 신뢰구간(신뢰구간제외 : errorbar=None)

sns.pointplot(data=iris, x='species', y='sepal_width', color='steelblue')

plt.show()
```

### 5) 회귀 그래프

```python
corr = iris.corr(numeric_only=True)
corr
```

```python
sns.heatmap(corr, vmin=-1, vmax=1, annot=True, fmt='.3f', cmap='Blues', cbar=False)
plt.show()
```

```python
np.ones_like(corr, dtype=bool) # corr과 구조가 같은 1로 채워진 배열
```

```python
mask = np.triu(np.ones_like(corr, dtype=bool))
sns.heatmap(corr, vmin=-1, vmax=1, annot=True, fmt='.3f', cmap='Blues', mask=mask)
plt.show()
```

```python
# 회귀그래프 : regplot(hue사용불가), lmplot(hue사용가능)

sns.regplot(data=iris, x='petal_length', y='petal_width')
plt.show()
```

```python
g = sns.lmplot(data=iris, x='petal_length', y='petal_width', hue='species',
              col='species')

g.fig.set_size_inches(10,3)
plt.show()
```

## 2.4 다중그래프를 위한 FacetGrid

```python
# 다중그래프

g = sns.FacetGrid(iris, col='species', hue='species')
g.map(sns.scatterplot, 'petal_length', 'petal_width')
plt.show()
```

```python
# 서브플롯

g = sns.relplot(iris, x='petal_length', y='petal_width', col='species', hue='species')
g.fig.set_size_inches(10,3)
sns.move_legend(g, "upper left", bbox_to_anchor=(0.92, 1))
plt.show()
```
